```python
import pandas as pd
import re
from scipy import stats

# Baca file Excel (ubah 'dataKasus-1.xlsx' dan '2022' sesuai dengan file kamu)
df = pd.read_excel('dataKasus-1.xlsx', sheet_name='2022')

# Fungsi untuk mengekstraksi angka dari string usia
def extract_number(text):
    # Memastikan teks diubah menjadi string terlebih dahulu lalu ambil angka
    match = re.search(r'\d+', str(text))
    return int(match.group()) if match else None  # Mengembalikan None jika tidak ada angka

# Terapkan fungsi ke kolom 'USIA'
df['USIA'] = df['USIA'].apply(extract_number)

# Menghitung Q1 dan Q3 untuk outlier
Q1 = df['USIA'].quantile(0.25)
Q3 = df['USIA'].quantile(0.75)
IQR = Q3 - Q1

# Mendefinisikan batas bawah dan atas untuk outlier
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Mendapatkan data tanpa outlier
df_no_outlier = df[(df['USIA'] >= lower_bound) & (df['USIA'] <= upper_bound)]

# Simpan hasil ke file Excel baru jika diperlukan
df_no_outlier.to_excel('cleaned_data.xlsx', index=False)

print(df_no_outlier)

```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Contoh data, pastikan sesuai dengan dataset asli
data = pd.DataFrame({
    'JARAK KELAHIRAN': ['anak pertama', '< 2 tahun', '> 2 tahun', '< 2 tahun', 'anak pertama', '> 2 tahun']
})

# Mengubah nilai kategori menjadi angka
category_mapping = {'anak pertama': 0, '< 2 tahun': 1, '> 2 tahun': 2}
data['JARAK KELAHIRAN'] = data['JARAK KELAHIRAN'].replace(category_mapping)

# Cek apakah ada nilai yang tidak valid (misalnya NaN)
print("Jumlah nilai NaN sebelum normalisasi:", data['JARAK KELAHIRAN'].isna().sum())

# Jika ada NaN, isi dengan median (atau kamu bisa menggunakan nilai lain sesuai kebutuhan)
data['JARAK KELAHIRAN'].fillna(data['JARAK KELAHIRAN'].median(), inplace=True)

# Normalisasi menggunakan Min-Max Scaler
scaler = MinMaxScaler()
data['JARAK KELAHIRAN Normalized'] = scaler.fit_transform(data[['JARAK KELAHIRAN']])

# Menampilkan hasil
print(data)

```
