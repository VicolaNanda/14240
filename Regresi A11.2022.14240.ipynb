```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Load data
file_path = '/content/produksiPadi.xlsx'
sheet_name = 'data training'
df = pd.read_excel(file_path, sheet_name=sheet_name)
```

```python
# Bersihkan data
df_cleaned = df[['Tahun', 'Produksi Padi(Ton)']].dropna()
df_cleaned.columns = ['Tahun', 'Produksi_Padi']
```

```python
# Definisikan variabel independen (X) dan dependen (y)
X = df_cleaned[['Tahun']]
y = df_cleaned['Produksi_Padi']
```

```python
# Split data menjadi training dan testing (70:30 split)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Buat model regresi linear
model = LinearRegression()
```

```python
# Latih model
model.fit(X_train, y_train)
```

```python
# Prediksi data uji
y_pred = model.predict(X_test)
```

```python
# Evaluasi model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
# Tampilkan hasil evaluasi
print("Mean Squared Error (MSE):", mse)
print("R-squared (R²):", r2)
print("Intercept:", model.intercept_)
print("Coefficient:", model.coef_)
```

```python
# Visualisasi hasil
plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, color='blue', label='Data Aktual')
plt.plot(X_test, y_pred, color='red', label='Prediksi (Regresi Linear)')
plt.title('Linear Regression - Produksi Padi')
plt.xlabel('Tahun')
plt.ylabel('Produksi Padi (Ton)')
plt.legend()
plt.show()
```
